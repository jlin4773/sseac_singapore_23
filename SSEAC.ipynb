```python
import pandas as pd
import requests
import folium
from folium.plugins import MarkerCluster

df = pd.read_csv('Downloads/Singapore_Postcodes.csv')

postcode_columns = ['Bishan', 'Ci Yuan', 'Kampong Glam', 'Online', '1', '2', '3']

base_url = 'http://www.mapquestapi.com/geocoding/v1/address'

api_key = 'y4y7Z4OhYe1SwXDNR0iDSHGTdeYXI0Zm'

marker_colors = {
    'Bishan': 'blue',
    'Ci Yuan': 'black',
    'Kampong Glam': 'pink',
    'Online': 'orange',
    '1': 'red',
    '2': 'white',
    '3': 'purple'
}

m = folium.Map(location=[1.3521, 103.8198], zoom_start=11)

marker_cluster = MarkerCluster().add_to(m)

geocoded_postcodes = set()

for index, row in df.iterrows():
    for col in postcode_columns:
        postcode = row[col]
        
        if pd.notnull(postcode) and postcode not in geocoded_postcodes:
            geocoded_postcodes.add(postcode)

            url = f'{base_url}?key={api_key}&location={postcode},Singapore'

            response = requests.get(url)
            
            result = response.json()
            lat = result['results'][0]['locations'][0]['latLng']['lat']
            lng = result['results'][0]['locations'][0]['latLng']['lng']
            
            if lat and lng:

                color = marker_colors[col]
                
                folium.Marker([lat, lng], icon=folium.Icon(color=color)).add_to(marker_cluster)

m.save('postcode_map.html')


m
```
